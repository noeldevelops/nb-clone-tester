## IMAGES
`Image(url)`outputs html image tag with `text/html` mimetype, not embedded

`Image(filename)` outputs base64 encoded image, embedded into doc

```python
# image/jpeg
# this renders base64 image/jpeg
Image(filename="lamb-chops_300x300.jpeg")
```

```python
# image/gif
# this renders base64 image/jpg
Image(filename="kermit.gif")
```

```python
# image/png
# this renders base64 image/jpeg
Image(filename="Noteable-Company-Logo.png")
```

```python
# image/svg+xml - works as expected
SVG(filename='Noteable-Company-Logo.svg')
```
